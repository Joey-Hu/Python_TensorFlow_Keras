```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
# 载入数据集
mnist = input_data.read_data_sets("../../data/mnist", one_hot=True)

# 定义每个批次大小(50?, 200?)
batch_size = 100

# 计算一共有多少个批次
n_batch = mnist.train.num_examples // batch_size

# 定义两个placeholder
x = tf.placeholder(tf.float32, [None, 784])
y = tf.placeholder(tf.float32, [None, 10])
keep_prob = tf.placeholder(tf.float32)

# 输入层
# 初始化服从正态分布的权值
Weights_L1 = tf.Variable(tf.truncated_normal([784, 1000], stddev=0.1))
biases_L1 = tf.Variable(tf.zeros([1000])+0.1)    
L1 = tf.nn.tanh(tf.matmul(x, Weights_L1) + biases_L1)
# dropout 计算有多少神经元不工作
L1_dropout = tf.nn.dropout(L1, keep_prob)

# 隐藏层
Weights_L2 = tf.Variable(tf.truncated_normal([1000, 1000], stddev=0.1))
biases_L2 = tf.Variable(tf.zeros([1000])+0.1)    
L2 = tf.nn.tanh(tf.matmul(L1, Weights_L2) + biases_L2)
# dropout 计算有多少神经元不工作
L2_dropout = tf.nn.dropout(L2, keep_prob)
                         
# 隐藏层
Weights_L3 = tf.Variable(tf.truncated_normal([1000, 1000], stddev=0.1)) 
biases_L3 = tf.Variable(tf.zeros([1000])+0.1)    
L3 = tf.nn.tanh(tf.matmul(L2, Weights_L3) + biases_L3)
# dropout 计算有多少神经元不工作
L3_dropout = tf.nn.dropout(L3, keep_prob)

                         
Weights_L4 = tf.Variable(tf.truncated_normal([1000, 10], stddev=0.1))   
biases_L4 = tf.Variable(tf.zeros([10])+0.1) 
prediction = tf.nn.softmax(tf.matmul(L3_dropout, Weights_L4) + biases_L4)     

# 交叉熵函数
loss =tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y, logits=prediction))

# 梯度下降法最小化代价函数（使用其他优化方式?）
train = tf.train.GradientDescentOptimizer(0.2).minimize(loss)

# 初始化变量 
init = tf.global_variables_initializer()

# 存放结果在一个布尔型列表中
# 首先让我们找出那些预测正确的标签。tf.argmax 是一个非常有用的函数，
# 它能给出某个tensor对象在某一维上的其数据最大值所在的索引值。由于标签向量是由0,1组成，因此最大值1所在的索引位置就是类别标签，
# 比如tf.argmax(y,1)返回的是模型对于任一输入x预测到的标签值，而 tf.argmax(y_,1) 代表正确的标签，
# 我们可以用 tf.equal 来检测我们的预测是否真实标签匹配(索引位置一样表示匹配)。
correct_prediction = tf.equal(tf.arg_max(y,1), tf.arg_max(prediction,1))

# 求准确率
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

# 执行会话
with tf.Session() as sess:
    sess.run(init)
    for epoch in range(21):    
        for batch in range(n_batch):
            batch_xs, batch_ys = mnist.train.next_batch(batch_size)
            sess.run(train, feed_dict={x: batch_xs, y: batch_ys, keep_prob:0.7})
        test_acc = sess.run(accuracy, feed_dict={x: mnist.test.images, y: mnist.test.labels, keep_prob:1.0})
        train_acc = sess.run(accuracy, feed_dict={x: mnist.train.images, y: mnist.train.labels, keep_prob:1.0})
        print("Iter" + str(epoch) + ",Testing Acc " + str(test_acc) + ",training Acc " + str(train_acc))
```
